```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler , normalize
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.utils import compute_sample_weight


```

```python

# 1. Separate features and labels first
features = df.iloc[:, 1:30]  # Skip 'Time' column
labels = df.iloc[:, 30]      # Target column

# 2. Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# 3. Normalize the features (row-wise L1 normalization)
features_normalized = normalize(features_scaled, norm="l1")

# 4. Assign to X and y
x = features_normalized
y = labels.values  # Convert to NumPy array
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

Build a Decision Tree Classifier model with Scikit-LearnÂ¶

```python
w_train = compute_sample_weight('balanced', y_train)
```

```python
decission_tree = DecisionTreeClassifier(max_depth=4, random_state=42)
decission_tree.fit(x_train, y_train, sample_weight=w_train)
y_pred = decission_tree.predict(x_test)
roc_auc = roc_auc_score(y_test, y_pred)
print(f"ROC AUC Score: {roc_auc:.4f}")
# Plotting the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 8))
plot_tree(decission_tree, filled=True, feature_names=df.columns[1:30], class_names=['Normal', 'Fraud'], rounded=True)
plt.title('Decision Tree Visualization')
plt.show()
```

SVM

```python
from sklearn.calibration import LinearSVC


svm = LinearSVC(class_weight='balanced', random_state=31,loss='hinge',fit_intercept = False)
svm.fit(x_train, y_train)
```

```python
y_pred_dt  = decission_tree.predict_proba(x_test)[:, 1]
```

Using these probabilities, we can evaluate the Area Under the Receiver Operating Characteristic Curve (ROC-AUC) score as a metric of model performance. The AUC-ROC score evaluates your model's ability to distinguish positive and negative classes considering all possible probability thresholds. The higher its value, the better the model is considered for separating the two classes of values.

```python
roc_auc_dt = roc_auc_score(y_test, y_pred_dt)
print('Decision Tree ROC-AUC score : {0:.3f}'.format(roc_auc_dt))
```

```python
y_pred_svm = svm.decision_function(x_test)
roc_auc_svm = roc_auc_score(y_test, y_pred_svm)
print('SVM ROC-AUC score : {0:.3f}'.format(roc_auc_svm))
```
